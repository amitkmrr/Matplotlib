```python
#matplotlib is a plotting library
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a=[25,32,45,67]
plt.plot(a)
plt.show()
```

```python
#plot x and y list
x=[23,45,6,56,7]
y=[234,897,673,78,56]
plt.title("Random diagram")
plt.xlabel("x-axis-value")
plt.ylabel("y-axis-value")
plt.plot(x,y,color="green",marker="D",markersize=20,linewidth=5,linestyle="dotted",alpha=0.5)
# plt.plot(x,y,'g+')
plt.show()
```

```python
#Axis label , legend, grid
days=[1,2,3,4,5,6,7]
mx_t=[50,51,52,34,48,47,49]
mn_t=[43,42,23,15,-1,9,10]
avg_t=[45,48,46,40,42,41,38]
plt.title("weather")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.plot(days,mx_t,label="max")
plt.plot(days,mn_t,label="min")
plt.plot(days,avg_t,label="avg")
#plt.legend
plt.legend(loc="best",shadow=True,fontsize="large")
plt.grid()
plt.show()
```

```python
#Bar Chart
company=["Google","Amazon","MSFT","Facebook"]
revenue=[90,136,89,29]
profit=[40,2,34,12]
xpos=np.arange(len(company))
plt.title("Revenue and Profit Bar")
plt.xlabel("US Tech Comapny")
plt.ylabel("Revenue and Profit(bn)")
plt.xticks(xpos,company)
plt.bar(xpos-0.2,revenue,label="revenue",width=0.4)
plt.bar(xpos+0.2,profit,label="profit",width=0.4)
#horizontal
# plt.barh(xpos-0.2,revenue,label="revenue")
# plt.barh(xpos+0.2,profit,label="profit")
plt.legend()
plt.show()
```

```python
#Histogram
blood_sugar=[115,85,90,150,140,88,93,112,135,80,77,82,129]
plt.hist(blood_sugar,bins=10,rwidth=0.95)
plt.show()
```

```python
blood_sugar_men=[115,85,90,150,140,88,93,112,135,80,77,82,129]
blood_sugar_woman=[111,65,80,130,40,98,93,102,105,118,57,112,139]
plt.hist([blood_sugar_men,blood_sugar_woman],bins=[80,100,125,150],color=["green","orange"],label=["Men","Woman"])
plt.legend()
plt.savefig("MyGenFie.pdf",bbox_inches="tight",pad_inches=2,transparent=True)
plt.show()
```

```python
#pie chart
exp_vals=[400,600,300,410,250]
exp_label=["Home Rent","Food","Phone/Internet","car","Others"]
plt.pie(exp_vals,labels=exp_label,radius=1.5,autopct='%0.1f%%',shadow=True)
plt.axis("equal")
plt.savefig("MyGenFie.png",bbox_inches="tight",pad_inches=2,transparent=False)
plt.show()
```
